```python
# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
'''

Estimation statistics, often referred to as simply "estimation," is a framework for data analysis that focuses on understanding the magnitude and 
uncertainty of effects rather than just presence or absence.

It uses a combination of:
Effect sizes :- Numerical values quantifying the size of an effect, like the difference between groups in an experiment.
Confidence intervals :- Ranges of values within which the true effect size is likely to fall, reflecting the uncertainty of the estimate.
Precision planning :-Determining the appropriate sample size needed to achieve desired levels of precision in the estimates.
Meta-analysis :- Combining data from multiple studies to arrive at more robust conclusions.

Here's a breakdown of the two key concepts in estimation:

1> Point Estimate : A single numerical value used to represent the most likely value of a population parameter based on a sample observation.Common examples include:
Mean: Average value in a sample (used for estimating population mean)
Proportion: Number of successes divided by the sample size (used for estimating population proportion)
Median: The "middle" value in a sorted list (alternative to mean for non-normal distributions)

2> Interval Estimate : Interval estimate: A range of values within which the true population parameter is likely to fall, with a specified level of confidence (usually 95%). 
This acknowledges the inherent uncertainty in estimates based on samples. For instance, a 95% confidence interval for the population mean might say "we are 95% confident that the true mean lies between X and Y."

Both point and interval estimates are crucial for understanding the results of a study. Point estimates give you a single value to represent the effect, while interval estimates reflect the uncertainty around that value.
```

```python
# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
'''
Hypothesis testing is a statistical method used to make inferences about population parameters based on a sample of data. It involves formulating a hypothesis about the population parameter,
collecting and analyzing data from a sample, and drawing conclusions about the validity of the hypothesis.
Here's a basic outline of the hypothesis testing process:

Formulate Hypotheses:
Null Hypothesis (H0): A statement that there is no significant difference or effect, often representing the status quo or a default assumption.
Alternative Hypothesis (H1 or Ha): A statement that contradicts the null hypothesis, suggesting a significant difference or effect.

Collect Data:
Gather a sample of data relevant to the hypothesis being tested.

Choose Significance Level (α):
The significance level (often denoted as α) is the probability of rejecting the null hypothesis when it is true. Commonly used values are 0.05, 0.01, or 0.10.

Conduct Statistical Test:
Use a statistical test (e.g., t-test, chi-square test, ANOVA) to analyze the data and determine the likelihood of observing the results under the null hypothesis.

Make a Decision:
Based on the test results, decide whether to reject the null hypothesis in favor of the alternative hypothesis, or fail to reject the null hypothesis.

Draw Conclusions:
Interpret the results in the context of the specific problem and make conclusions about the population based on the sample data.

----- > Importance of Hypothesis Testing:

Informed Decision Making:
Hypothesis testing provides a structured framework for making decisions based on data, allowing researchers, analysts, and decision-makers to draw conclusions about population parameters.

Scientific Research:
In scientific research, hypothesis testing is crucial for validating or refuting scientific theories. It helps researchers make meaningful contributions to their fields by providing evidence for or against hypotheses.

Quality Control and Process Improvement:
Industries use hypothesis testing to assess the quality of products or processes. It helps identify improvements or changes that can enhance efficiency and effectiveness.

Medical and Pharmaceutical Research:
Hypothesis testing is widely used in medical and pharmaceutical research to evaluate the effectiveness of treatments, drugs, and interventions.

Policy and Decision-Making:
Governments and organizations use hypothesis testing to assess the impact of policies and decisions, ensuring that choices are evidence-based.

Risk Management:
In finance and business, hypothesis testing is employed to assess risks, make investment decisions, and evaluate the impact of changes in business strategies.
'''
```

```python
# Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
'''
Null hypothesis: The average weight of male college students is equal to the average weight of female college students.

Alternative hypothesis: The average weight of male college students is greater than the average weight of female college students.

This hypothesis is directional, meaning it specifies the direction of the expected difference (male weight being greater).
```

```python
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
  '''
  Performs a two-sample t-test on the difference between two population means.

  Args:
      sample1 (list): A list of data points from the first population.
      sample2 (list): A list of data points from the second population.
      alpha (float, optional): The significance level for the test. Defaults to 0.05.

  Returns:
      dict: A dictionary containing the following information:
          - statistic (float): The test statistic (t-value).
          - pvalue (float): The p-value of the test.
          - reject_null (bool): Whether to reject the null hypothesis (True if p-value < alpha).
  '''

  # Check if samples have equal variances
  f_statistic, pvalue_f = stats.f_oneway(sample1, sample2)
  equal_variances = pvalue_f > alpha

  # Perform the appropriate t-test based on equal variances
  if equal_variances:
    t_statistic, pvalue = stats.ttest_ind(sample1, sample2, equal_var=True)
  else:
    t_statistic, pvalue = stats.ttest_ind(sample1, sample2, equal_var=False)

  # Reject null hypothesis if p-value is less than significance level
  reject_null = pvalue < alpha

  return {
      "statistic": t_statistic,
      "pvalue": pvalue,
      "reject_null": reject_null,
  }

# Example usage
sample1 = [70, 75, 80, 82, 78]  # Replace with your data
sample2 = [65, 68, 72, 69, 71]  # Replace with your data

results = two_sample_t_test(sample1, sample2)

print("Test statistic:", results["statistic"])
print("p-value:", results["pvalue"])
print("Reject null hypothesis:", results["reject_null"])

if results["reject_null"]:
  print("There is significant evidence to reject the null hypothesis. The average weight of sample 1 is likely different from sample 2.")
else:
  print("There is not enough evidence to reject the null hypothesis. The average weight of sample 1 might be similar to sample 2.")
```

```python
# Q6: What is a null and alternative hypothesis? Give some examples

'''
In statistical hypothesis testing, null and alternative hypotheses are two competing statements about a population parameter, like a mean, proportion, or variance.
They play a crucial role in guiding the analysis and interpreting the results.


Null hypothesis (H₀):
Represents the "no effect" scenario, usually assuming the status quo or no difference between groups.
It's typically denoted by H₀ and considered the default position until evidence suggests otherwise.
You aim to reject the null hypothesis to support your research question.

Alternative hypothesis (H₁):
States the specific prediction or expected effect you're investigating.
It's denoted by H₁ and expresses the alternative explanation to the null hypothesis.
You fail to reject the null hypothesis if evidence doesn't strongly support H₁.

Examples:

Research question: Does studying with flashcards improve exam scores compared to no studying?
H₀: There is no difference in exam scores between students who study with flashcards and those who don't. (No effect)
H₁: Students who study with flashcards score higher on exams than those who don't. (Positive effect)

'''
```

```python

# Q7: Write down the steps involved in hypothesis testing.
'''
1. Formulate the hypotheses:
State the null hypothesis (H₀): This represents the "no effect" scenario, usually assuming the status quo or no difference between groups.
State the alternative hypothesis (H₁): This is your specific prediction or expected effect you're investigating.

2. Choose an appropriate statistical test:
Select a test based on the type of data (e.g., numerical, categorical), sample size, and research question. Common tests include t-tests, ANOVA, chi-square tests, and correlations.

3. Define the level of significance (α):
This is the probability of rejecting the null hypothesis when it's actually true (Type I error). It's typically set to 0.05 (5%).

4. Collect data:
Gather a representative sample from the population of interest using appropriate sampling methods.

5. Perform the statistical test:
Apply the chosen test to your data, following the specific instructions and assumptions of the test.

6. Calculate the test statistic and p-value:
The test statistic measures the "extremeness" of the observed data under the null hypothesis.
The p-value is the probability of obtaining a test statistic as extreme or more extreme than the observed one, assuming the null hypothesis is true.

7. Make a decision:
Compare the p-value to the level of significance (α).
If p-value ≤ α: Reject the null hypothesis. There is evidence to support the alternative hypothesis.
If p-value > α: Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.
```

```python
# Q8. Define p-value and explain its significance in hypothesis testing.
'''

The p-value is a fundamental concept in hypothesis testing, reflecting the probability of observing results at 
least as extreme as those obtained, assuming the null hypothesis is true.

Significance levels and decision-making:
-> P-values are typically compared to a pre-defined significance level (usually α = 0.05) which represents 
   the acceptable probability of making a Type I error (rejecting a true null hypothesis). 
-> P-value <= α: This suggests the observed results are unlikely to be due to chance alone, providing evidence to reject the null hypothesis. You can tentatively support the alternative hypothesis.
-> P-value > α: This implies the observed results could be explained by chance, and there's not enough evidence to reject the null hypothesis. You fail to reject the null hypothesis, but it doesn't necessarily mean the null hypothesis is true.

Important points to remember:

P-value is not the probability of the null hypothesis being true: It only signifies the likelihood of observing your specific results under the null hypothesis.
Lower p-value doesn't guarantee a larger effect size: A small p-value with a small effect size might not be practically significant.
Consider sample size: Smaller samples can lead to larger p-values even with real effects, requiring careful interpretation.
P-value alone doesn't tell the whole story: Analyze effect size, confidence intervals, and broader context for stronger conclusions.
```

```python
# Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

# Set degrees of freedom
df = 10

# Generate x-values for the plot
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)

# Calculate the probability density function values
y = t.pdf(x, df)

# Create the plot
plt.plot(x, y, label="Student's t-distribution, df=10")

# Customize plot elements
plt.xlabel("t-values")
plt.ylabel("Probability Density")
plt.title("Student's t-Distribution with df=10")
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
'''Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.'''

import numpy as np
from scipy.stats import ttest_ind

def two_sample_ttest(sample1, sample2, alpha=0.05):
    """
    Performs a two-sample t-test for independent samples.

    Args:
        sample1 (list): A list of numerical values for the first sample.
        sample2 (list): A list of numerical values for the second sample.
        alpha (float, optional): The significance level for the test. Defaults to 0.05.

    Returns:
        dict: A dictionary containing the following information:
            - statistic (float): The t-statistic value.
            - pvalue (float): The p-value of the test.
            - reject_null (bool): Whether to reject the null hypothesis (True if p-value < alpha).
    """

    # Check for equal sample sizes
    if len(sample1) != len(sample2):
        raise ValueError("Samples must have equal sizes for a two-sample t-test.")

    # Perform the t-test
    t_statistic, pvalue = ttest_ind(sample1, sample2)

    # Reject null hypothesis if p-value is less than significance level
    reject_null = pvalue < alpha

    return {
        "statistic": t_statistic,
        "pvalue": pvalue,
        "reject_null": reject_null,
    }

# Example usage
sample1 = [70, 75, 80, 82, 78]  # Replace with your first sample data
sample2 = [65, 68, 72, 69, 71]  # Replace with your second sample data

results = two_sample_ttest(sample1, sample2)

print("Test statistic:", results["statistic"])
print("p-value:", results["pvalue"])
print("Reject null hypothesis:", results["reject_null"])

if results["reject_null"]:
    print("There is significant evidence to reject the null hypothesis. The population means are likely different.")
else:
    print("There is not enough evidence to reject the null hypothesis. The population means might be equal.")
```

```python
 # Q11: What is Student’s t distribution? When to use the t-Distribution
'''
Student's t-distribution explained:
The Student's t-distribution, also known as the t-distribution, is a probability distribution similar to the familiar bell-shaped normal distribution. 
However, it has fatter tails, meaning it's more likely to have extreme values compared to the normal distribution.

Key features of the t-distribution:

-> Shape: Bell-shaped, symmetrical around zero, but with fatter tails than the normal distribution.
-> Parameters: One parameter called degrees of freedom (df), which influences the spread of the distribution. Higher degrees of freedom lead to a shape closer to the normal distribution.
-> Applications: Primarily used in hypothesis testing and constructing confidence intervals, especially when dealing with small samples or unknown population variances. Examples include:
-> Student's t-test: Comparing the means of two independent groups.
-> Paired t-test: Comparing the means of two related groups.
-> Confidence intervals for population means: Estimating the range within which the true population mean is likely to lie.'''
```

```python

# Q12: What is t-statistic? State the formula for t-statistic.
'''
There are different variations of the t-statistic formula depending on the specific type of t-test being conducted. Here are some common ones:

One-sample t-test:

This compares the mean of a sample to a hypothesized population mean:

t = (x̄ - μ) / (s / √n)
where:

x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation
n is the sample size
Independent samples t-test:

This compares the means of two independent groups:

t = (x̄₁ - x̄₂) / √ [(s₁² / n₁) + (s₂² / n₂)]
where:

x̄₁ and x̄₂ are the means of the two groups
s₁² and s₂² are the variances of the two groups
n₁ and n₂ are the sample sizes of the two groups
'''
```
